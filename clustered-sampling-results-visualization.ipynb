```python
def get_n_clients(arr):
    ans = 0
    for i in range(10):
        ans += max(arr[i*10:i*10+10])
    return ans
def extract_metrics(loss_path,acc_path,clients_path):
    with open(loss_path,'rb') as f:
        loss = pickle.load(f)
    with open(acc_path,'rb') as f:
        acc = pickle.load(f)
    with open(clients_path,'rb') as f:
        clients = pickle.load(f)
    d = {
        "loss_mean": [np.mean(loss[i]) for i in range(1,len(loss),2)],
        "loss_std": [np.std(loss[i]) for i in range(1,len(loss),2)],
        "acc_mean": [np.mean(acc[i]) for i in range(1,len(acc),2)],
        "acc_std": [np.std(acc[i]) for i in range(1,len(acc),2)],
        'sampled_clients': [get_n_clients(clients[i]) for i in range(1,len(clients),1)],
    }
    return d
```
